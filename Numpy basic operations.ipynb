```python
# 4. Operazioni elementari Dato un array numerico: calcola il quadrato di ogni elemento, aggiungi 5 a ogni elemento, moltiplica ogni elemento per 2

import numpy as np
import random

list_number  = [random.randint(0,10) for i in range(0,10)]
my_array = np.array(list_number)

# Array iniziale
print("The initial array is:", my_array)

# Calcola il quadrato di ogni elemento
print("The array with the square of each element is:", my_array**2)
# Aggiungi 5 a ogni elemento
print("The array where we added 5 to each element is", my_array+5)
#Moltiplica ogni elemento per 2
print("The arry where each element is multiplied by 2 is:", my_array*2)
```

```python
# 5. Statistiche di base Dato un array di numeri: calcola media, minimo, massimo, deviazione standard

import numpy as np
import random

#Array with pseudo ramdom integer numbers
my_array = np.array([random.randint(0,100) for index in range(0,1000)])

#We employ Numpy methods to compute
print("The Mean is:",  np.mean(my_array))
print("The Standard deviation is:", np.std(my_array))
print("The minimum value is:", np.min(my_array), "while the maximum value is:", np.max(my_array))
```

```python
# 6. Indicizzazione e slicing Dato un array: estrai i primi 5 elementi, estrai gli ultimi 3 estrai gli elementi in posizione pari

import numpy as np
import random

data = np.array([random.randint(0,10) for index in range(0,10)])
print("We consider the following array:", data)

# The first five elements are
print("The list of the first five elements is:",list(data[0:5:1]))
# The list 3 elements are
print("The list of the last three elements is:", data[-3::1])
# The elements with an even indez are
print("Elemens with an even index:", data[1::2])

```

```python
# 7. Operazioni condizionali. Dato un array: sostituisci tutti i valori negativi con 0, conta quanti valori sono maggiori di una soglia data

import random
import numpy as np

data = np.array([random.randint(-10,10) for index in range(0,20)])

# Print the initial array
print("We consider the following array of random integer numbers:", data)

# Change all the negative values with zero
data[data < 0] = 0
print("The new array with only positive or equal to zero values is:", data)

# Count how many values are bigger than a number
number = int(input("Insert a number between 0 and 10:"))

data_bigger = data > number
print("The number of element bigger than", number, "is", data_bigger.sum())


```

```python
# 8. Maschere booleane. Dato un array: crea una maschera per i valori ≥ 0.5 usa la maschera per estrarre solo quei valori

import random
import numpy as np

values = np.array([random.random() for i in range(0,100)])
print("We cosnider the array:", values)

# Boolean mask
mask = values >= 0.5
wanted_values = values[mask]

print("We keep the values:", wanted_values)
print("This amounts to keep:", len(wanted_values), "elements")
```

```python
# 9. Operazioni tra array. Dati due array della stessa lunghezza: somma elemento per elemento prodotto elemento per elemento 
# differenza assoluta tra gli elementi

import numpy as np
import random

length = 10
vec1 = np.array([random.randint(0,10) for i in range(0,length)])
vec2 = np.array([random.randint(0,10) for j in range(0,length)])

print("The first vector is:", vec1)
print("The second vector is:", vec2)

# The sum 
print("The sum of the two vectors is:", vec1+vec2)
# The product element by element is
print("Product:", vec1*vec2)
# The absolute difference between the elements is
print("The absolute difference between the elements is:", np.abs(vec1-vec2))

```

```python
# 10. Mini data-cleaning con NumPy Dato un array che contiene valori validi e np.nan:
# calcola la media ignorando i nan, sostituisci i nan con la media. restituisci l’array pulito

import numpy as np

# Initial array
data = np.array([12.5, np.nan, 7.8, 10.0, np.nan, 3.2, 8.9, np.nan, 15.4])

mask = np.isnan(data)
numerical_data = data[~mask]
#Compute the mean
avg = np.mean(numerical_data)
print("The mean of the numerical data is:", avg)
# New array nan |-> avg
data[mask] = avg
print("the new arry is:", data)
```
